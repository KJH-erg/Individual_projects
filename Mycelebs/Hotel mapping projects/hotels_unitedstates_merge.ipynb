```python
id_hotels=[]
id_matching=[]
Country_eng=[]
Country_ko=[]
city=[]
HotelName_hotels=[]
HotelAddress_hotels=[]
Latitude_hotels=[]
Longitude_hotels=[]
HotelName=[]
HotelAdress=[]
Latitude=[]
Longitude=[] 
sub_country_list = []
sub_city_list = []
where = []
def compare_name(name1, name2):
    if (name1 in name2) or (name2 in name1):
        return 1
    else:
        return 0
######################################################################################################################
for i in tqdm(range(len(sub))):  #i는 sub의 index
    hotels['is_contain'] = 0
    hotels['is_contain'] = hotels['HotelFileName'].apply(compare_name, name2=sub['sub_HotelName'].loc[i])
    sub_match_id_list = hotels['HotelID'][hotels['is_contain'] == 1].to_list()
    for k in sub_match_id_list:  # 각 id를 활용해 도시가 같은지 비교
        
        if format(sub['sub_Latitude'].loc[i],".2f") == format(hotels['Latitude'][hotels['HotelID']==k].values[0],".2f") and format(sub['sub_Longitude'].loc[i],".2f") == format(hotels['Longitude'][hotels['HotelID']==k].values[0],".2f"):

        #if sub['sub_CityName'].loc[i] == hotels['PlaceFileName'][hotels['HotelID']==k].values[0] and format(sub['sub_Latitude'].loc[i],".2f") == format(hotels['Latitude'][hotels['HotelID']==k].values[0],".2f") and format(sub['sub_Longitude'].loc[i],".2f") == format(hotels['Longitude'][hotels['HotelID']==k].values[0],".2f"):
            # gg의 인덱스인 i와 hotels의 id인 k를 활용 
            hc_id = k
            hc_CountryName = hotels['CountryName'][hotels['HotelID'] == hc_id ].values[0]
            hc_PlaceName = hotels['PlaceName'][hotels['HotelID'] == hc_id ].values[0]
            hc_HotelName = hotels['HotelName'][hotels['HotelID'] == hc_id ].values[0]
            hc_HotelAddress = hotels['HotelAddress'][hotels['HotelID'] == hc_id ].values[0]
            hc_Latitude = hotels['Latitude'][hotels['HotelID'] == hc_id ].values[0]
            hc_Longitude = hotels['Longitude'][hotels['HotelID'] == hc_id ].values[0]
            sub_id = sub['sub_HotelID'].loc[i]
            sub_HotelName = sub['sub_HotelName'][sub["sub_HotelID"] == sub_id].values[0]
            sub_HotelAdress = sub['sub_HotelAddress'][sub["sub_HotelID"] == sub_id].values[0]
            sub_Latitude = sub['sub_Latitude'][sub["sub_HotelID"] == sub_id].values[0]
            sub_Longitude = sub['sub_Longitude'][sub["sub_HotelID"] == sub_id].values[0]
            sub_country = sub['sub_CountryName'][sub["sub_HotelID"] == sub_id].values[0]
            sub_city = sub['sub_CityName'][sub["sub_HotelID"] == sub_id].values[0]
            sub_from = sub['from'][sub['sub_HotelID']== sub_id].values[0]
            id_hotels.append(hc_id)
            Country_ko.append(hc_CountryName)
            city.append(hc_PlaceName)
            HotelName_hotels.append(hc_HotelName)
            HotelAddress_hotels.append(hc_HotelAddress)
            Latitude_hotels.append(hc_Latitude)
            Longitude_hotels.append(hc_Longitude)
            id_matching.append(sub_id)
            HotelName.append(sub_HotelName)
            HotelAdress.append(sub_HotelAdress)
            Latitude.append(sub_Latitude)
            Longitude.append(sub_Longitude) 
            sub_country_list.append(sub_country) 
            sub_city_list.append(sub_city)
            where.append(sub_from)
        else:
            continue
result = pd.DataFrame(list(zip(id_matching, HotelName, HotelAdress, sub_country_list, sub_city_list, Latitude, Longitude,
                               where, id_hotels, HotelName_hotels, HotelAddress_hotels, Country_ko, city, 
                               Latitude_hotels, Longitude_hotels)),
                          columns=['sub_HotelID', 'sub_HotelName', 'sub_HotelAddress', 'sub_CountryName',
                                   'sub_CityName', 'sub_Latitude', 'sub_Longitude',
                                   'from', 'HotelID', 'HotelName', 'HotelAddress', 'CountryName',
                                   'PlaceName', 'Latitude', 'Longitude'])
result
```
