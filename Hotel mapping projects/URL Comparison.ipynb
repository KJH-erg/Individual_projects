```python
import time
from selenium import webdriver
import pandas as pd
from haversine import haversine
import tqdm


df = pd.read_excel('rate1.xlsx' , 'Sheet1')
url_matching = [] # list for the result
df['url_matching'] = 0 # column for the result
driver=webdriver.Chrome('./chromedriver')
```

```python
for i in range(len(df)):
    search1 = '"'+ str(df.iloc[i]['sub_HotelName']) +'"'+ ' and ' + str(df.iloc[i]['sub_HotelAddress'])
    search2 = '"'+ str(df.iloc[i]['HotelName']) + '"'+ ' and ' + str(df.iloc[i]['HotelAddress'])
    print(search1)
    print()
    print(search2)
    driver.get("https://www.google.co.kr/maps")
    driver.implicitly_wait(1)
    elem = driver.find_element_by_id("searchboxinput")
    elem.send_keys(search1)
    elem = driver.find_element_by_id('searchbox-searchbutton')
    elem.click()
    time.sleep(5)
    url1 = driver.current_url
    driver.get('https://www.google.co.kr/maps')
    driver.implicitly_wait(1)
    elem = driver.find_element_by_id('searchboxinput')
    elem.send_keys(search2)
    elem = driver.find_element_by_id('searchbox-searchbutton')
    elem.click()
    time.sleep(5)
    url2 = driver.current_url
    if url1 == url2:
        url_matching.append(1) # if url is eqaul add 1 to list
        print("1")
    else:
        url_matching.append(0) # if url is not eqaul add 0 to list
        print("0")
driver.close()
```

```python
#list value to colum
df ["url_matching"] = url_matching
```

```python
#export
df.to_excel('result.xlsx')
```
