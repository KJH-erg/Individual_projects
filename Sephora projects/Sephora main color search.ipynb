```python
import pandas as pd
import time
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager # webdriver-manager 패키지 다운로드
from fake_useragent import UserAgent
from tqdm import tqdm
from bs4 import BeautifulSoup
import requests





startno = 3500
endno = 4000
Data = pd.read_excel('./glamai_file.xlsx', 'not_mapped') # 데이터 파일경로
Data = Data[startno:endno]


Ori = pd.read_excel('./glamai_file.xlsx', 'sephora_keyword')
Ori = Ori[0:len(Ori)]
Data["Main_c"] = 0
Data["Main_cna"] = 0
Data["redundancy"] = 0

```

```python
i =0
j = 0
for i in tqdm(range(0,len(Data))):
    
    for j in range(len(Ori)):
        if (Ori.loc[j,'product_code'] == Data.loc[i+startno,"PRODUCT_CODE"]):
            Data.loc[i+startno,"Main_c"] = Ori.loc[j , 'main_sku']

        
```

```python
ua = UserAgent(verify_ssl=False)
userAgent = ua.random
print(userAgent)
```

```python
options = webdriver.ChromeOptions()
options.add_argument(f'user-agent={userAgent}')
from selenium.webdriver.common.alert import Alert
from selenium.webdriver.common.keys import Keys

drv = webdriver.Chrome(ChromeDriverManager().install(),chrome_options=options)
url_check = list()
drv.get("https://www.sephora.com/")
time.sleep(20)
url = 0


for i in range(len(Data)):
    if ((Data.loc[i+startno, 'PRODUCT_CODE'])!=(Data.loc[i+startno+1, 'PRODUCT_CODE'])):
        search =str(Data.loc[i+startno, 'PRODUCT_CODE'])+ ' ' + str(Data.iloc[i]['Main_c'])
        print(search)
        elem = drv.find_element_by_id("site_search_input")
        elem.send_keys(search)
        elem.submit()
        req = requests.get(drv.current_url)
        source = req.text
        soup = BeautifulSoup(source, 'html.parser')
        try:
            no = (soup.select("body > div.css-o44is > div.css-18qzcvp > main > div:nth-child(2) > div:nth-child(1) > div > div.css-1kk8dps.e65zztl0 > div.css-1jg4myg.e65zztl0 > span"))[0].text
            Data.loc[i+startno,"Main_cna"] = no
        except:
            Data.loc[i+startno,"Main_cna"] = 0
    else:
        print("")
drv.close()             
```
